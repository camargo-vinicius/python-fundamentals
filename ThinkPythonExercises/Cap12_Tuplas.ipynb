##### Exercício 12.2
Escreva uma programa que leia words.txt e imprima todos os conjuntos de palavras que são anagramas

```python
def is_anagram(str1: str, str2: str) -> bool:

    # para uma ser anagrama da outra, elas devem ter exatamente as mesmas letras
    # podemos entao ordenar as strings e ver se elas sao iguais
    return sorted(str1.lower()) == sorted(str2.lower())

string1, string2 = 'estudo', 'duetos'

print(f'{string2} eh anagrama de {string1}: {is_anagram(string1, string2)}')
```

```python
with open('../words.txt', 'r') as file:
    words = file.read().split('\n')[:-1]

# cria um dict onde as letras das palavras são armazenadas como tuplas
dict_letters = {tuple(sorted(word)): [] for word in words}

for word in words:

    # converte cada palavra do loop numa tupla para checar sua existencia como chave do dict
    word_tuple = tuple(sorted(word))

    if word_tuple in dict_letters:
        dict_letters[word_tuple].append(word)

# filtra o dicionario apenas onde os valores (que são listas) possuem mais de um elemento
# pois aí sim sao anagramas
dict_filtered = {k: v for k, v in dict_letters.items() if len(v) > 1}

print(dict_filtered)
    
```
