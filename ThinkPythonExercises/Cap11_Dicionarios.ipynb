##### Exercício 11.2
Reescreva a função invert_dict usando o método setdefault de dicionários

```python
# montando um dicionario com um histograma das letras de uma palavra para depois inverte-lo
# entao na chave temos a letra e o valor a qtd de vezes q a letra repete na palavra

def invert_dict(dct: dict) -> dict:
    dict_inverse = {}

    for k, v in dct.items():
        
        if v not in dict_inverse:

            # seta como default na chave v o valor [k] (lista de um item)
            dict_inverse.setdefault(v, [k])

        else:
            # se a chave já existir, acrescenta o item lista
            dict_inverse[v].append(k)
        
    return dict_inverse

palavra = 'brontosauro'

hist = {letra: palavra.count(letra) for letra in palavra}
inverse_hist = invert_dict(hist)

print(f'Dicionario original eh: {hist}\n')
print(f'Dicionario invertido eh: {inverse_hist}\n')
```

##### 11.4. Reescreva a função has_duplicate feita no cap 10, porém usando um dicionário

```python
# código original: já usando dicionario rsrsrs
def has_duplicates(lista: list) -> bool:
    
    dict_count = {num: lista.count(num) for num in lista}
    
    for v in dict_count.values():
        
        if v > 1:
            return True
    
    return False

lista_de_listas = [[1, 2, 2, 3, 4, 5], ['a', 'b', 'c', 'd', 'c'], ['a', 'b', 'c'], [10, 11, 12, 14]]

for item in lista_de_listas:
    print(f'Lista {item} contém duplicatas: {has_duplicates(item)}')
```

```python
def has_duplicates(lista: list) -> bool:
    
    dict_count = {num: lista.count(num) > 1 for num in lista}

    # deixando um pouco mais pythonico usando funcao any nos valores do dict, que sao booleanos
    # se algum deles tiver a contagem > 1, isto é, possui duplicatas, retorna True
    return any(dict_count.values())

lista_de_listas = [[1, 2, 2, 3, 4, 5], ['a', 'b', 'c', 'd', 'c'], ['a', 'b', 'c'], [10, 11, 12, 14]]

for item in lista_de_listas:
    print(f'Lista {item} contém duplicatas: {has_duplicates(item)}')
```

##### Exercício 11.5
Escreva uma funcao que leia uma lista de palavras e retorne todos os pares rotacionados

```python
import string

def rotate_word(word: str, step: int = 5) -> str:
    
    # string vazia para receber a nova string formada
    new_word = ''
    
    for letter in word:
        
        # verifica se é caractere especial. Se for, não rotaciona a letra
        if letter not in string.punctuation:

            # define o start de acord com o case da letra
            start = ord('a') if letter.islower() else ord('A')

            # formula para calcular o index
            index =  start + (ord(letter) - start + step) % 26

        else:
            index = ord(letter)
        
        new_word += chr(index)

    return new_word
```
